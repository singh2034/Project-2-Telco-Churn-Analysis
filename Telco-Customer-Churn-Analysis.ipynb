#Replacing the blanks with 0 as the tenure is 0 and no total charges are recorded

```python
df['TotalCharges'] = df['TotalCharges'].replace(" ", "0")
df['TotalCharges'] = df['TotalCharges'].astype("float")
```

```python
df.describe()
```

```python
#check for duplicate
df.duplicated().sum()
```

```python
#coversion of data 0 for senior citizen so that it can be readable
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
df.head()
```

```python
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of customers by Churn", fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({"Churn" : "count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned customers", fontsize = 10)
plt.show()
```

1) From the given pie chart we can conclude that 26.54% of the customers are churned out.
2) Now let's explore the reasong behind it.

```python
plt.figure(figsize = (5,3))
sns.countplot( x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender", fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (5,3))
ax = sns.countplot( x = "SeniorCitizen", data = df)
plt.title("Count of Cusotmer by Senior Citizen", fontsize = 10)
plt.show()
```

```python
# Calculate percentages
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# Adjust figure size for better visualization
fig, ax = plt.subplots(figsize=(6, 6))

# Plot the bars
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])

# Add percentage labels on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}', ha="center", va="center")

# Add title, labels, and legend
plt.title("Churn by Senior Citizen")
plt.xlabel('SeniorCitizen')
plt.ylabel("Percentage (%)")
plt.xticks(rotation=0)
plt.legend(title='Churn', loc='upper right')

# Show the plot
plt.show()
```

Comparatively a greater percentage of people in senior citizen have churned.

```python
#On the basis of tenure
plt.figure(figsize = (9,4))
sns.histplot(x = "tenure", data = df, bins = (72), hue="Churn")
plt.show()
```

People who have used our services for long time have stayed and people who have used our services for 1 or 2 months have churned.

#now on the above data we want to create the subplots where we will use ChatGPT for the faster work(we should know what we want from it).

```python
# Columns for which you want to create countplots
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Define the layout of the subplots (3 rows, 3 columns)
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))
axes = axes.flatten()  # Flatten the 2D array of axes for easy iteration

# Loop through each column and plot
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, hue=col, ax=axes[i], palette='Set2', legend=False)
    axes[i].set_title(f'Countplot for {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Remove any empty subplots if the number of columns doesn't fit perfectly
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

# Adjust spacing between plots
plt.tight_layout()
plt.show()
```
